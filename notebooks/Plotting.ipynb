```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = [15, 8]
```

```python
df0 = pd.read_csv("scores_dqn.csv")
df1 = pd.read_csv("scores-target-network.csv")
df2 = pd.read_csv("scores_ddqn.csv")
```

```python
plt.plot(df0["Episodes"], df0["Scores"], label='DQN')
plt.plot(df1["Episodes"], df1["Scores"], label='DQN-Target-Network')
plt.plot(df2["Episodes"], df2["Scores"], label='DDQN')
plt.legend()
plt.show()
```

```python
df3 = pd.read_csv("scores_ddqn-v1-150k.csv")
df4 = pd.read_csv("scores_ddqn-v2-150k.csv")
plt.plot(df0["Episodes"], df0["Scores"], label='DQN')
plt.plot(df1["Episodes"], df1["Scores"], label='DQN-Target-Network')
plt.plot(df2["Episodes"], df2["Scores"], label='DDQN')
plt.plot(df3["Episodes"], df3["Scores"], label='DDQN-v1')
plt.plot(df4["Episodes"], df4["Scores"], label='DDQN-v2')
plt.legend()
plt.show()
```
