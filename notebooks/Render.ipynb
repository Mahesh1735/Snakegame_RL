```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

fig = plt.figure()
data = env.reset()
im = plt.imshow(data, cmap="YlGn")

def init():
    data, b, c = env.step(np.random.randint(4))
    if c:
        data = env.reset()
    im.set_data(data)

def animate(i):
    data, b, c = env.step(np.random.randint(4))
    im.set_data(data)
    return im

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=50, interval=200, repeat = True)
```
